## 1: Coding neurons and layers

## 1.1: Neuron with 3 inputs

```python
inputs = [7, 3, 5]
weights = [0.5, -0.6, 0.8]
bias = 1.2

# Calculating the output of the neuron
output = (inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2] * weights[2] + bias)
print(f"Output of neuron with 3 inputs: {output}")
```

## 1.2: Neuron with 4 inputs

```python
inputs = [6, 2, 4, 3.5]
weights = [0.3, -1.0, 0.4, 1.3]
bias = 1.6

# Calculating the output of the neuron
output = (
    inputs[0] * weights[0] +
    inputs[1] * weights[1] +
    inputs[2] * weights[2] +
    inputs[3] * weights[3] +
    bias
)
print(f"Output of neuron with 4 inputs: {output}")
```

## 1.3: First layer

```python
inputs = [1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],
 [0.5, -0.91, 0.26, -0.5],
 [-0.26, -0.27, 0.17, 0.87]]

weights1 = weights[0] #LIST OF WEIGHTS ASSOCIATED WITH 1ST NEURON : W11, W12, W13, W14
weights2 = weights[1] #LIST OF WEIGHTS ASSOCIATED WITH 2ND NEURON : W21, W22, W23, W24
weights3 = weights[2] #LIST OF WEIGHTS ASSOCIATED WITH 3RD NEURON : W31, W32, W33, W34

inputs = [2.0, 3.5, 1.5, 2.2]

weights = [
    [0.2, 0.4, -0.5, 0.7],   # Weights for Neuron 1
    [-0.1, 0.9, 0.3, -0.3],  # Weights for Neuron 2
    [0.5, -0.7, 0.2, 0.4]    # Weights for Neuron 3
]

weights1 = weights[0] # W11, W12, W13, W14
weights2 = weights[1] # W21, W22, W23, W24
weights3 = weights[2] # W31, W32, W33, W34

biases = [1.4, 2.3, 0.9]

# Calculating the output of the layer
outputs = [
    # Neuron 1
    inputs[0] * weights[0][0] +
    inputs[1] * weights[0][1] +
    inputs[2] * weights[0][2] +
    inputs[3] * weights[0][3] + biases[0],
    # Neuron 2
    inputs[0] * weights[1][0] +
    inputs[1] * weights[1][1] +
    inputs[2] * weights[1][2] +
    inputs[3] * weights[1][3] + biases[1],
    # Neuron 3
    inputs[0] * weights[2][0] +
    inputs[1] * weights[2][1] +
    inputs[2] * weights[2][2] +
    inputs[3] * weights[2][3] + biases[2]
]

print(f"Outputs of the layer: {outputs}")
```

```python
inputs = [2.0, 3.5, 1.5, 2.2]

weights = [
    [0.2, 0.4, -0.5, 0.7],   # Weights for Neuron 1
    [-0.1, 0.9, 0.3, -0.3],  # Weights for Neuron 2
    [0.5, -0.7, 0.2, 0.4]    # Weights for Neuron 3
]

biases = [1.4, 2.3, 0.9]

# Calculating the output of the layer using a loop
layer_outputs = []

# For each neuron
for neuron_weights, neuron_bias in zip(weights, biases):
    # Zeroed output of given neuron
    neuron_output = 0
    # For each input and weight to the neuron
    for n_input, weight in zip(inputs, neuron_weights):
        # Multiply this input by associated weight
        # and add to the neuron's output variable
        neuron_output += n_input * weight
    # Add bias
    neuron_output += neuron_bias
    # Put neuron's result to the layer's output list
    layer_outputs.append(neuron_output)

print(f"Outputs of the layer: {layer_outputs}")
```

# 2: Using numpy:

## 2.1: Single neuron

```python
import numpy as np

inputs = [2.0, 3.5, -1.5, 4.0]
weights = [0.5, -0.3, 0.8, -0.2]
bias = 1.0
output = np.dot(weights, inputs) + bias

print(output)
```

## 2.2: Layer of neurons using numpy

```python
inputs = [1.5, -2.0, 3.3, 0.7]

# Weights for each neuron
weights = [
    [0.4, -0.2, 0.5, 1.2],
    [0.6, -0.8, 0.3, -0.7],
    [-0.3, 0.9, -0.4, 0.5]
]

# Biases for each neuron
biases = [1.0, -1.5, 0.8]

# Calculate the layer outputs
layer_outputs = np.dot(weights, inputs) + biases

print(layer_outputs)
```

## 2.3: Layer of neurons and batch of data using numpy

```python
inputs = [
    [2.0, -1.0, 4.0, 1.5],
    [3.0, 0.5, -2.0, 3.0],
    [-2.5, 3.0, 1.5, -1.2]
]

# Weights for each neuron
weights = [
    [0.3, -0.7, 0.6, 1.1],
    [0.8, -1.2, 0.4, -0.6],
    [-0.4, 0.5, -0.3, 0.9]
]

# Biases for each neuron
biases = [1.5, -0.5, 2.0]

# WE CAN'T TRANSPOSE LISTS IN PYTHON, SO WE HAVE THE CONVERT THE WEIGHTS MATRIX INTO AN ARRAY FIRST
outputs = np.dot(inputs, np.array(weights).T) + biases

print(outputs)
```

```python
inputs = [
    [2.5, -1.0, 4.0, 1.5], 
    [3.0, 0.5, -2.5, 3.5], 
    [-2.0, 3.5, 2.0, -1.5]
]

# Weights for each neuron
weights = [
    [0.4, -0.3, 0.6, 1.2],
    [0.7, -1.0, 0.3, -0.8],
    [-0.5, 0.6, -0.2, 1.0]
]

# Biases for each neuron
biases = [1.5, -0.5, 2.5]

inputs_array = np.array(inputs)
weights_array = np.array(weights)
biases_array = np.array(biases)

outputs = np.dot(inputs_array, weights_array.T) + biases_array
print(outputs)
```

# 3: Multiple nn layers

```python
import numpy as np

# Input samples
inputs = [
    [2.0, -1.0, 4.0, 1.5],
    [3.5, 0.5, -2.5, 3.0],
    [-2.0, 3.5, 2.0, -1.5]
]

# Weights for the first layer
weights = [
    [0.4, -0.3, 0.6, 1.2],
    [0.7, -1.0, 0.3, -0.8],
    [-0.5, 0.6, -0.2, 1.0]
]

# Biases for the first layer
biases = [1.5, -0.5, 2.5]

# Weights for the second layer
weights2 = [
    [0.2, -0.14, 0.5],
    [-0.5, 0.12, -0.33],
    [-0.44, 0.73, -0.13]
]

# Biases for the second layer
biases2 = [-1, 2, -0.5]

# Convert lists to numpy arrays
inputs_array = np.array(inputs)
weights_array = np.array(weights)
biases_array = np.array(biases)
weights2_array = np.array(weights2)
biases2_array = np.array(biases2)

# Calculate the output of the first layer
layer1_outputs = np.dot(inputs_array, weights_array.T) + biases_array

# Calculate the output of the second layer
layer2_outputs = np.dot(layer1_outputs, weights2_array.T) + biases2_array

print(layer2_outputs)
```
